```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("IHR Geo-Visual.csv")
#df
```

```python
df.columns
```

```python
sn. barplot(data = df, x = "State", y = "Total_Women_works")
plt.xticks(rotation=90)
plt.show()
```

```python
sn. barplot(data = df, x = "Total_Male_works", y = "State")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sn.histplot(df['Main Workers Rural Persons'], bins=20, kde=True, color='blue', label='Total Workers')
sn.histplot(df['Main Workers Rural Males'], bins=20, kde=True, color='green', label='Males')
sn.histplot(df['Main Workers Rural Females'], bins=20, kde=True, color='orange', label='Females')
plt.legend()
plt.xlabel('Number of Workers')
plt.ylabel('Counts')
plt.title('Distribution of Main Workers in Rural Areas')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sn.histplot(df[['Main Workers Urban Persons', 'Main Workers Urban Males', 'Main Workers Urban Females']], bins=20, kde=True)
plt.legend(['Total Workers', 'Males', 'Females'])
plt.xlabel('Number of Workers')
plt.ylabel('Counts')
plt.title('Distribution of Main Workers in Urban Areas')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sn.lineplot(x='Main Workers Rural Females', y='Marginal Workers Rural Females', data=df, color='blue', marker='o')
plt.xlabel('Main Rural Females Workers', fontsize=12)
plt.ylabel('Marginal Rural Females Workers', fontsize=12)
plt.title('Relationship between Main and Marginal Rural Female Workers', fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the sum of workers for the pie chart
total_main_workers = df['Main Workers Total Persons'].sum()
total_marginal_workers = df['Marginal Workers Total Persons'].sum()

# Data for pie chart
pie_data = [total_main_workers, total_marginal_workers]
labels = ['Main Workers', 'Marginal Workers']

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(pie_data, labels=labels, autopct='%1.1f%%', startangle=90)

# Adding a title
plt.title('Distribution of Main Workers vs Marginal Workers - Total', fontsize=14)

# Equal aspect ratio ensures the pie is drawn as a circle
plt.axis('equal')

# Display the pie chart
plt.show()
```

```python
# Box plots for numerical columns
sn.boxplot(data=df, orient="h")
plt.show()
```
